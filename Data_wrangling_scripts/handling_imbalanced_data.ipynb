```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/kingcarlos/DIVINE_LUMINARY/UCSD_Course/Capstone_Project/Project_datasets/wordnet_data/cleaned_wordnet_data_label_encoded.csv'
data = pd.read_csv(file_path)

# Analyze the distribution of the "Part of Speech" column
pos_counts = data['Part of Speech'].value_counts()

# Visualize the POS tag distribution
pos_counts.plot(kind='bar', title='POS Tag Distribution', xlabel='POS Tags', ylabel='Count')
plt.show()
```
